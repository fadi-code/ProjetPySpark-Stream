```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace, col, length, expr, date_format, when, current_timestamp
import psycopg2
from datetime import datetime
import pytz

# 1. Cr√©er une session Spark
print("Cr√©ation de la session Spark...")
spark = SparkSession.builder \
    .appName("PostgresToSpark") \
    .config("spark.jars", "/usr/local/spark/jars/postgresql-42.3.9.jar") \
    .getOrCreate()

# 2. Configurer les param√®tres de connexion
print("Configuration des param√®tres de connexion JDBC pour PostgreSQL...")
jdbc_url = "jdbc:postgresql://postgres:5432/DB_Mastodon"  # Remplacez 'postgres' et 'DB_Mastodon' par vos valeurs
properties = {
    "user": "fadi",        # Votre nom d'utilisateur PostgreSQL
    "password": "fadi",    # Votre mot de passe PostgreSQL
    "driver": "org.postgresql.Driver"  # Driver JDBC PostgreSQL
}

# 3. Lire les donn√©es depuis PostgreSQL
print("Lecture des donn√©es depuis la table 'Mostodon_SILVER' dans PostgreSQL...")
df = spark.read.jdbc(url=jdbc_url, table="Mostodon_SILVER", properties=properties)
print(f"Nombre de lignes lues : {df.count()}")

# 4. Nettoyer la colonne 'content' des balises HTML
print("Suppression des balises HTML dans la colonne 'content'...")
df_cleaned = df.withColumn("content", regexp_replace(col("content"), "<[^>]+>", ""))

# 5. Calcul des nouvelles colonnes
print("Calcul des nouvelles colonnes...")
df_with_calculations = df_cleaned.withColumn("engagement_score", 
    (col("favourites_count") * 0.5) + 
    (col("reblogs_count") * 1.0) + 
    (col("replies_count") * 1.5)
).withColumn("content_length", length(col("content"))) \
 .withColumn("hour_of_day", expr("HOUR(created_at)")) \
 .withColumn("day_of_week", date_format(col("created_at"), "EEEE")) \
 .withColumn("hashtags", expr("regexp_extract_all(content, '#\\w+')")) \
 .withColumn("contains_link", when(col("content").rlike("http[s]?://"), True).otherwise(False)) \
 .withColumn("days_since_posted", (current_timestamp().cast("long") - col("created_at").cast("long")) / (3600 * 24)) \
 .withColumn("normalized_engagement", 
    (col("favourites_count") + col("reblogs_count") + col("replies_count")) /
    ((current_timestamp().cast("long") - col("created_at").cast("long")) / 3600)
)

# 6. Afficher les r√©sultats calcul√©s
df_with_calculations.show()

# 7. Connexion √† PostgreSQL via psycopg2
print("Connexion √† PostgreSQL avec psycopg2...")
conn = psycopg2.connect(
    host="postgres",  # Changez ceci si n√©cessaire
    database="DB_Mastodon",
    user="fadi",
    password="fadi"
)
cursor = conn.cursor()

# 8. Cr√©ation de la table 'Mostodon_GOLD' si elle n'existe pas d√©j√†
print("Cr√©ation de la table 'Mostodon_GOLD' si elle n'existe pas d√©j√†...")
create_table_query = '''
CREATE TABLE IF NOT EXISTS Mostodon_GOLD (
    id BIGINT PRIMARY KEY,
    username TEXT NOT NULL,
    display_name TEXT NOT NULL,
    content TEXT NOT NULL,
    created_at TIMESTAMP NOT NULL,
    favourites_count INT NOT NULL,
    reblogs_count INT NOT NULL,
    replies_count INT NOT NULL,
    engagement_score FLOAT NOT NULL,
    content_length INT NOT NULL,
    hour_of_day INT NOT NULL,
    day_of_week TEXT NOT NULL,
    hashtags TEXT[],
    contains_link BOOLEAN NOT NULL,
    days_since_posted INT NOT NULL,
    normalized_engagement FLOAT NOT NULL
);
'''
cursor.execute(create_table_query)
conn.commit()

# 9. Insertion des donn√©es calcul√©es dans la table 'Mostodon_GOLD'
print("Insertion des donn√©es calcul√©es dans la table 'Mostodon_GOLD' via Spark JDBC...")
df_with_calculations.write.jdbc(url=jdbc_url, table="Mostodon_GOLD", mode="append", properties=properties)
print("Insertion termin√©e.")

# 10. Compter le nombre total d'entr√©es dans la table 'Mostodon_GOLD'
print("Calcul du nombre total d'entr√©es dans la table 'Mostodon_GOLD'...")
cursor.execute("SELECT COUNT(*) FROM Mostodon_GOLD;")
count = cursor.fetchone()[0]  # R√©cup√©rer le r√©sultat du COUNT
print(f"Nombre total d'entr√©es dans la table 'Mostodon_GOLD' : {count}")

# 11. Fermer le curseur et la connexion PostgreSQL
cursor.close()
conn.close()
print("Connexion √† PostgreSQL ferm√©e.")

# 12. Fermer la session Spark
print("Fermeture de la session Spark...")
spark.stop()
print("Session Spark ferm√©e.")
```
